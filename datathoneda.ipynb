```python
transfact_features = transfact.drop(['frgn_curr_code', 'transaction_type', 'current_account_nbr', 'transaction_date', 'posting_date', 'adj_orgn_tran_dt', 'first_purchase_ind'], axis=1)
```

```python
X = transfact_features
y = transfact_features.transaction_amt
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('slope: ', model.coef_)
print('intercept: ', model.intercept_)
```
